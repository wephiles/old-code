# 保存评论

## 导入相关数据包

```python
import os
```

```python
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
from selenium import webdriver
import random
```

```python
import csv, codecs
```

```python
import solution_class
```

```python
path = Service(
        '../ChromeDriver/chromedriver.exe')  # 浏览器驱动程序地址
driver = webdriver.Chrome(service=path)
driver.maximize_window()  # 最大化窗口
driver.implicitly_wait(10)  # 等待时间超过10s时，抛出异常
```

## 抓取评论

```python
def get_comment(url: str = None) -> list[str]:
    """获取一门课程的所有评论
    
    参数：
    -------------------
    url:课程评论页面的网址
    """
    comment_list = []
    # 打开页面
    pro = solution_class.GetText(driver=driver)  # 实例化类
    pro.open_url(url=url)

    comment_button = pro.driver.find_element(By.ID, 'review-tag-button')
    comment_button.click()
    time.sleep(1)
    comment_list = pro.get_course_evaluation(pro)
    return comment_list
# get_comment('https://www.icourse163.org/course/SUDA-1001752241?from=searchPage&outVendor=zw_mooc_pcssjg_')
```

## 保存到文件

```python
def to_file(url: str = None) -> None:
    """把评论都写进文件中
    
    参数：
    -------------------
    url:课程评论页网址
    """
    pro = solution_class.GetText(driver=driver)  # 实例化类
    comment_list = []
    comment_list = get_comment(url)
    
    course_name = pro.get_course_name(pro)
    
    file_path = r'../result/' + course_name + '.txt'
    file_path_csv = r'../result/' + course_name + '.csv'
    
    if os.path.exists(file_path) :
        file_path = r'../result/' + course_name + str(random.randint(0, 30)) + '.txt'
        
    if os.path.exists(file_path_csv) :
        file_path_csv = r'../result/' + course_name + str(random.randint(0, 30)) + '.csv'
        
    with open(file_path, 'w', encoding='utf-8') as fp:
        for line in comment_list:
            if isinstance(line, int):
                line = str(line)
            fp.write(line)
            fp.write('\n')
    
    with open(file_path_csv, 'w', encoding='utf-8') as fp:
        for line in comment_list:
            if isinstance(line, int):
                line = str(line)
            fp.write(line)
            fp.write('\n')
    
    print('done')
    pro.driver.close()
    return None
```

```python
url = 'https://www.icourse163.org/course/NJUPT-1003219004?from=searchPage&outVendor=zw_mooc_pcssjg_'
to_file(url)
```
