# Pandas数据排序

Series的排序：  
Series.sort_values(ascending=True, inplace=False)  
参数说明：  
- ascending，默认为True升序排序，为False降序排序
- inplace，是否修改原始的Series

DataFrame的排序：
DataFrame.sort_values(by, ascending=True, inplace=False)
参数说明：

- by，字符串或者List<字符串>，单列排序或者多列排序
- ascending：bool或者List，升序还是降序，如果是list那么对应于by的多列
- inplace：是否修改原始dataframe

```python
import pandas as pd

file_path = './data/beijing_tianqi/beijing_tianqi_2018.csv'

df = pd.read_csv(file_path)
df
```

```python
df.loc[:, 'bWendu'] = df['bWendu'].str.replace('℃', '').astype('int32')
df.loc[:, 'yWendu'] = df['yWendu'].str.replace('℃', '').astype('int32')
```

```python
df
```

# Series的排序

```python
df['aqi'].sort_values()
```

```python
df['aqi'].sort_values(ascending=False)
```

```python
df['tianqi'].sort_values()
```

# Dataframe的排序

## 单列排序

```python
df
```

```python
df.sort_values(by='aqi',)
```

```python
df.sort_values(by='aqi', ascending=False)
```

## 多列排序

```python
# 按照空气质量等级、最高温度排序 默认升序

df.sort_values(by=['aqiLevel', 'bWendu']).head(10)
```

```python
# 两个字段都是降序
df.sort_values(by=['aqiLevel', 'bWendu'], ascending=False).head(10)
```

```python
# 分别指定降序和升序
df.sort_values(by=['aqiLevel', 'bWendu'], ascending=[True, False]).head(10)
```
